### Controlling execution with the  *continue* and *pass* statement



> Contoh 

* menggunakan function input()

```python
value = input("Ketiklah sebuah kata yang kurang dari 6 karakter: ")

number = 1

for kata in value:
  print ("Kata ", number, " adalah ",kata )
  number += 1
  if number == 4:
    print ("Maaf, huruf ini tidak akan diproses!")
    continue
```

> Contoh menggunakan Continue

```python
profile = ['champion', 'loyal', 'potential loyalist', 'promising', 'new customers',
    'about to sleep', 'lost']
counter= 0

for level in profile:
  if level == 'promising':
    continue
    print ('Profil ini tidak akan diproses')
  print ('Berikut adalah tingkatan profil, ', counter, " adalah ", level)
  counter += 1
    
```

> Contoh menggunakan Pass

```python
profile = ['champion', 'loyal', 'potential loyalist', 'promising', 'new customers',
    'about to sleep', 'lost']
counter= 0

for level in profile:
  if level == 'promising':
    pass
    print ('Profil ini tidak akan diproses')
  print ('Berikut adalah tingkatan profil, ', counter, " adalah ", level)
  counter += 1
    
```

### Nested for and while
> Contoh

Membuat tabel perkalian.



```python
# Buat kedua variabel
X = 1
Y = 1 
print ('{:>4}'.format(' '), end= ' ')


for X in range(1, 11):
  print('{:>4}'.format(X), end=' ')
print()

for X in range(1,11):
  print('{:>4}'.format(X), end=' ')
  while Y <= 10:
    print('{:>4}'.format(X * Y), end=' ')
    Y+=1
  print()
  Y=1
```

## Nested Functions


```python
# function di dalam function 

def berbicara (text):
  def berbisik (t):
    return t.lower() + ' .... '
  return berbisik (text)

berbicara ('Selamat Malam semuanya')
```

```python
# inner function tidak dapat dipanggil di luar function utama

berbisik (aaa)
```

```python
# inner function juga tidak dapat diteruskan sebagai argumen di function lain

berbicara (berbisik)
```

```python
# Untu dapat mengakses inner function di luar function berbicara maka kita gunakan 
# return untuk mendapatkan inner function kembali ke caller dari function parent

def ambil_berbicara_func (volume):
  def berbisik (text):
    return text.lower() + '...'
  def yell (text):
    return text.upper() + '!'
  if volume > 0.5:
    return yell
  else:
    return berbisik
```

```python
# kita coba panggil function ambil_berbicara_func

ambil_berbicara_func (0.4)
```

```python
bicara_func = ambil_berbicara_func (0.7)
bicara_func('halo, apakah sudah bisa ?')
```

## Lambdas 

```python
# define function menggunakan def keyword
def penjumlahan1 (x,y):
  return x+y 

#panggil function dengan memasukkan argumen
penjumlahan1 (10, 20)
```

```python
# membuat function dengan lambda keyword

# buat variabel
penjumlahan2 = lambda x,y: x + y 

# panggil function penjumlahan 2 dengan memasukkan argumen
penjumlahan2 (10,20)
```

```python
# lambda as a single expression

(lambda x, y: x+y) (10,20) 
# lambda expression tanpa mendefinisikan "penjumlahan2" function, dan dapat memanggil saat itu juga menggunakan argumen 10,20
```
